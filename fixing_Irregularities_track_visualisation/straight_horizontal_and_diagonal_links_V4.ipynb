```python
def straight_diagonal(path1 : "Where json file is located", x1, x2, y1, y2, x_fix, y_fix, n): 
    with open(path1) as data_file:    
        data = json.load(data_file) 
    df = pd.DataFrame(data['nodes']) 
    node_id= []
    node_X = []
    node_Y = []
    node_name = []
    a = []
    b = []
    c = []
    d = []
    list_node = (list(zip(df['id'],df['x'],df['y'],df['name'])))
    for values in list_node:
        ID = values[0]
        x = values[1]
        y = values[2]
        Name = values[3]
        if y1<y2: 
#             if (x1-1) < x < (x2+1):
            if x1 <= x <= x2:
                if  y1 < y < y2:
                    a.append(ID)
                    b.append(x)
                    c.append(y)
                    d.append(Name)
                else:
                    node_id.append(ID)
                    node_X.append(x)
                    node_Y.append(y)
                    node_name.append(Name)
            else:
                node_id.append(ID)
                node_X.append(x)
                node_Y.append(y)
                node_name.append(Name)
        else:
#             if (x1-1) < x < (x2+1):
            if x1 < x <= x2:
                if  y2 < y < y1:
                    a.append(ID)
                    b.append(x)
                    c.append(y)
                    d.append(Name)
                else:
                    node_id.append(ID)
                    node_X.append(x)
                    node_Y.append(y)
                    node_name.append(Name)
            else:
                node_id.append(ID)
                node_X.append(x)
                node_Y.append(y)
                node_name.append(Name)
    
    diag_fix = list(zip(a,b,c,d))
    diag_fix = sorted(diag_fix, key = lambda t: t[1])
    for i in diag_fix:
        if y1<y2:
            x = x_fix + n
            y = y_fix + n
            n += 0.02
            node_id.append(i[0])
            node_X.append(x)
            node_Y.append(y)
            node_name.append(i[3]) 
        else:
            x = x_fix + n
            y = y_fix - n
            n += 0.02
            node_id.append(i[0])
            node_X.append(x)
            node_Y.append(y)
            node_name.append(i[3]) 
            
        
    df_final = pd.DataFrame()
    df_final.insert(0,"id",node_id)
    df_final.insert(1,"x",node_X)
    df_final.insert(2,"y",node_Y)
    df_final.insert(3,"name",node_name)
    df_final = df_final.to_json(orient = 'records')
    json_object = json.loads(df_final)
    data['nodes'] = json_object
    with open(path1, 'w') as json_file:
        json.dump(data, json_file)

    
```
